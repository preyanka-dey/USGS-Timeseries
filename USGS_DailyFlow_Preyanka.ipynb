### *Lets, download stream flow data from some USGS stations for Chickamauga Watershed, TN*

```python
sites = ['03566535', '03567500', '03567340','03566700']  # USGS gage ids 
start = '2017-01-01'
end = '2021-01-01'
service = 'dv'
# Request our data.
data = hf.NWIS(sites, service, start, end, file='usgs1.parquet') # csv should also work, but encoder need be specified 
data # Verify that the data request went fine.
```

```python
data1=pd.read_parquet('usgs1.parquet')
data1.head()
```

### *Let's only extract the discharge data & write the data into a csv file*

```python
Q= data.df('00060') # '00060' is for discharge
Q.to_csv('Discharge.csv', index=True)
```

```python
Q.head()
```

```python
Q.columns=['N Chick at Mile ST','S Chick at Ringold', 'W Chick at Lakeview', 'S Chick']
Q.head()
```

### *Plotting daily discharge*

```python
ax=Q.plot(figsize=(10,5))
ax.set(xlabel="Date",
       ylabel="Discharge (cfs)",
       title="Daily Discharge")
```

### *Plotting weekly total discharge*

```python
ax=Q.resample('w').sum().plot(figsize=(10,5)) # potting weekly total discharge data
ax.set(xlabel="Date",
       ylabel="Discharge (cfs)",
       title="Weekly Total Discharge")
```

### *Plotting monthly total discharge*

```python
ax=Q.resample('m').sum().plot(figsize=(10,5))  # monthly data 
ax.set(xlabel="Date",
       ylabel="Discharge (cfs)",
       title="Monthly Total Discharge")
```

### *Observing monthly pattern of discharge over 4 years*

```python
ax=Q.groupby(Q.index.month).mean().plot(figsize=(10,5));
ax.set(xlabel="Date",
       ylabel="Discharge (cfs)",
       title="Monthly Discharge Pattern")
ax.set_xlim([1, 12])
ax.set_xticks(range(1,12))
```

Discharge is higher in the first quarter of the year and slowly decreases in the late summer. As discharge is a result of precipitation, it should also follow the same pattern. Let's plot the Precipation data available for the first three stations and check the pattern. 

```python
P= data.df('00045') # extract precipitation data
P.columns=["N Chick at Mile ST","S Chick at Ringold ", 'W Chick at Lakeview']
P.to_csv('precip.csv', index=True)  # write the precipitation data in a csv file for later use
ax=P.plot(figsize=(10,5))
ax.set(xlabel="Date",
       ylabel="Precipitation (in)",
       title="Daily Total Precipitation")
```

Check Precipitation Pattern across the year

```python
P.groupby(P.index.month).mean().plot(figsize=(10,5));
```

So precipitation is high during the winter and decreases slowly during summer months 

Let's calculate the total discharge in South Chickamauga watershed, West Chick is inside South Chick watershed

```python
Q.loc[:,'Total South Chick']= Q['S Chick at Ringold']+Q['W Chick at Lakeview']+Q['S Chick']
```

```python
Q.head()
```

Now check the rolling mean, every discharge point is sum of previous 365 days discharge 

```python
ax=Q.resample('D').sum().rolling(365).sum().plot(figsize=(10,5)) # yearly rolling mean 
ax.set_ylim(0,None);
ax.set(xlabel="Date",
       ylabel="Discharge(cfs)",
       title="Yearly Rolling Stream Flow")
```

```python
df=Q
df.loc[:,'Month']=df.index.month
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(2,2,figsize=(12,12), sharey=False)
boxplot = df.boxplot(column=['N Chick at Mile ST', 'S Chick at Ringold','W Chick at Lakeview','S Chick'], 
                     by=['Month'],showfliers=False, ax=ax1)
fig.suptitle('Monthly FLow Pattern of Chickamauga Watershed Gages')
```

```python
type(df.index)
```

```python
dfm=df.resample('m').mean()
dfm.head()
```

Now, let's analysis the seasonality of USGS station data located near Ringgold in South Chickamauga watershed.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from matplotlib import pylab
from pylab import *
pylab.rcParams['figure.figsize'] = (10, 20)
result = seasonal_decompose(dfm['S Chick at Ringold'], model='additive')
result.plot()
plt.show()
```

A repeating yearly seasonal pattern is clearly visible in the seasonal plot.

```python
dfm2=dfm[['S Chick']]
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(2,1,figsize=(10,10))
sm.graphics.tsa.plot_acf(dfm2.squeeze(), lags=29,ax=ax1[0])
sm.graphics.tsa.plot_pacf(dfm2.squeeze(), lags = 29, ax=ax1[1])
plt.show()
```

Normally a time series is linearly related to a lagged version of itself. The coefficient of correlation between two values in a time series is called the autocorrelation function (ACF). The above acf and pacf plots represent the degree of similarity between a given time series and a lagged version of itself over successive time intervals. It can be seen that each month issimilar to the same month the previous year (lag 12 months).These plots can help to identify the seasonal pattern in the dataset. From these, we can get the idea of ARIMA model components (p, q) which will be discussed in a different tutorial.
